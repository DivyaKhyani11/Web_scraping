```python
import requests
import re
from bs4 import BeautifulSoup

api_key = "Unique Api KEY" #edited the name of API KEY for security reasons
location = "countryIN"
cse_id = "60728459df67743c7" #custom search engine id
query = "best Chinese restaurants Bangalore"
radius = 500

# Constructing the URL for the Google Custom Search API request
url = f"https://www.googleapis.com/customsearch/v1?q={query}&key={api_key}&cx={cse_id}"

# Defining parameters for the API request
params = {
    "cr": location, # Filter results from the specified country
    "key": api_key,
    "cx": cse_id,
    "q":query,
    "radius":radius
}

# Sending a GET request to the API and retrieving the response
response = requests.get(url, params=params)
# Parsing the JSON response data
data = response.json()
```

```python
#checking if the "data" variable contain the Parsed JSON response
data
```

```python
# Initializing an empty array to store the links from the API response
restaurants_links = []

# Iterating through each item in the 'items' list of the API response data
for i in data['items']:
    restaurants_links.append(i['link']) # Append the 'link' value from the current item to the 'restaurants_links' array

print(restaurants_links)
```

```python
restaurant_final_list = []
for link in restaurants_links:
  response = requests.get(link)
  if response.status_code == 200:
    response.raise_for_status()  # Check for successful response
    soup = BeautifulSoup(response.content, 'html.parser')  # Parse the HTML content
    if link == "https://curlytales.com/best-chinese-restaurants-in-bangalore/":
      list_of_res = []
      res=soup.find_all('h2')
      for r in res:
        # Define a regular expression pattern to match lines starting with a number
        pattern = r'^\d+\.\s+(.*)$'
        # Find all matches using re.findall
        matches = re.findall(pattern, r.text.strip(), re.MULTILINE)
        list_of_res.append(matches)
      restaurant_final_list.append([item for sublist in list_of_res for item in sublist])

    elif link == "https://www.gqindia.com/live-well/content/craving-for-chinese-food-in-bengaluru-check-out-this-list-of-best-chinese-restaurants-in-bangalore":
      restaurant_name_elements = soup.find_all('div', {'role': 'heading', 'aria-level': '3', 'class': 'heading-h3'})
      restaurant_final_list.append([element.get_text() for element in restaurant_name_elements])

    elif link == "https://lbb.in/bangalore/budget-chinese-bangalore/":
      restaurant_name_elements1 = soup.find_all('h2', class_='styled-components__StyledMultiSectionTitle-sc-w4s7c5-17')
      restaurant_final_list.append([element.get_text() for element in restaurant_name_elements1])

    elif link == "https://www.popxo.com/2020/01/best-chinese-restaurants-in-bangalore/":
      restaurant_name_elements2 = soup.find_all('h3')
      restaurant_final_list.append([element.get_text() for element in restaurant_name_elements2])

    elif link == "https://dineout.co.in/bangalore-restaurants/east-bangalore/whitefield/chinese-cuisine":
      restaurant_name_elements3 = soup.find_all('a', class_='restnt-name ellipsis')  # Find all <a> elements with the specified class
      restaurant_final_list.append([element.get_text() for element in restaurant_name_elements3])

    elif link == "https://magicpin.in/india/Bangalore/Nagawara/Restaurant/Chinese/":
      restaurant_name_elements4 = soup.find_all('h2', class_='merchant-name')
      restaurant_final_list.append([element.find('a').get_text(strip=True) for element in restaurant_name_elements4])

    elif link == "https://magicpin.in/india/Bangalore/Electronic-City/Restaurant/Chinese/":
      restaurant_name_elements5 = soup.find_all('h2', class_='merchant-name')
      restaurant_final_list.append([element.find('a').get_text(strip=True) for element in restaurant_name_elements5])
    else:
      continue
```

```python
restaurant_final_list_flat = [item for sublist in restaurant_final_list for item in sublist]
print(restaurant_final_list_flat,len(restaurant_final_list_flat))
```

```python
#removing duplicates from 'restaurant_final_list_flat'
test_list = list(set(restaurant_final_list_flat))
print(len(test_list))
```

```python
import csv

# Define the CSV file name
csv_file_name = 'restaurant_names.csv'
# new_column_data = ['Rating 4.5', 'Rating 3.8', 'Rating 4.2']

# Write the combined list to a CSV file
with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Best Chinese Restaurants in Bangalore'])  # Write header
    csv_writer.writerows([[name] for name in restaurant_final_list_flat])  # Write data


print("CSV file created:", csv_file_name)
```
